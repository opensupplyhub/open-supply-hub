```python
from splink import Linker, DuckDBAPI
import pandas as pd

db_api = DuckDBAPI()
```

```python
locs_df = pd.read_csv("records.csv")
locs_df
```

```python
loc_df = pd.read_csv("record.csv")
loc_df['os_id'] = ''

loc_df
```

```python
import json

with open("record_linkage_model_beta.json", "r") as file:
    record_linkage_model = json.load(file)

linker = Linker(
    locs_df,
    settings=record_linkage_model,
    db_api=db_api,
)
```

```python
df_inc = linker.inference.find_matches_to_new_records(
    loc_df, blocking_rules=[]
).as_pandas_dataframe()
df_inc.sort_values("match_weight", ascending=False)
df_inc.to_csv("record_linkage_output.csv", index=False)
# linker.inference.predict(threshold_match_probability=0.2).as_pandas_dataframe().sort_values(
#     "match_weight", ascending=False
# )
```
